```python
# Importando bibliotecas
import pandas as pd
```

```python
# Lendo o arquivo
df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx', skiprows=3, skip_footer=3)

# Arruma nome das colunas
df.columns = df.columns.str.strip()
```

```python
# Analisando o cabeçalho
df.head()
```

```python
# Análise do footer
df.tail()
```

```python
# Converter valor para número
df['Valor Bilhete'] = pd.to_numeric(df['Valor Bilhete'])

# 10 Bilhetes mais caros
df[['Nome do órgão superior', 'Valor Bilhete']].nlargest(10, 'Valor Bilhete')
```

```python
# Órgãos que mais emitiram passagem aéreas
qnt_passagens = df.groupby(['Nome do órgão superior']).size() \
                                      .reset_index(name='Quantidade de Passagens') \
                                      .sort_values(['Quantidade de Passagens'], ascending=False)\

qnt_passagens
```

```python
# importa bibliotecas de gráficos
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
# Estatísticas Relacionadas ao valor do bilhete

summary = df['Valor Bilhete'].describe()
print(summary)
```

```python
# Órgãos que mais gastaram com passagem aéreas

valor_passagem = df.groupby(['Nome do órgão superior'])['Valor Bilhete'].sum() \
                                                       .reset_index(name='Valor') \
                                                       .sort_values(['Valor'], ascending=False)
                             
valor_passagem
```

```python
# Gastos por Cia Aerea

cias = df['Companhia Aérea']

cia_valores = df.groupby(['Companhia Aérea'])['Valor Bilhete'].sum() \
                                                              .reset_index(name='Valor') \
                                                              .sort_values(['Valor'], ascending=False)
cia_valores
```

```python
# Gráfico das Cia que mais receberam dinheiro

plt.barh(cia_valores['Companhia Aérea'], cia_valores['Valor'])
plt.title('Gastos por Cia Área')
plt.xlabel('milhões de Reais')
plt.ylabel('Nome da Cia Aérea')
plt.show()
```

```python
# Top 10 Órgãos que mais gastaram

plt.barh(valor_passagem['Nome do órgão superior'].head(10), valor_passagem['Valor'].head(10)/1000000)
plt.title('Gastos por Órgão')
plt.xlabel('milhões de Reais')
plt.ylabel('Nome do Órgão')
plt.show()
```
